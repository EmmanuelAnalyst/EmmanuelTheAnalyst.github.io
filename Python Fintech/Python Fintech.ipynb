```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
fintech = pd.read_csv('C:/Users/EMMANUEL UDOH/Fintech.csv')
```

```python
fintech.head()
```

```python
fintech.dtypes
```

```python
#This is to check for null values in each column

for col in fintech.columns:
    pct_missing = np.mean(fintech[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
fintech.describe()
```

```python
pd.set_option("display.max_rows", None)  #This shows all rows and columns
```

```python
fintech.sort_values(by=['row_num'], inplace=False, ascending=True)  #The dataset has been sorted by row_num
```

```python
fintech['payment_type'].describe()  #This describes all the the details in the payment_type column
```

```python
sns.boxplot(x=fintech['payment_type'], y=fintech['reward_rate'], data = fintech)
```

```python
#The distplot below show the overall distribution of reward.rate from the data

res = sns.distplot(fintech.reward_rate, bins=10, color = 'g')
res
```

```python
#The lineplot shows the continuous relationship between the deposits and the withdrawal from different payment type.

plt.figure(figsize=(20,10))
plt.title('Deposits and withdrawal by payment type')
res = sns.lineplot(x='deposits', y='withdrawal', data = fintech, hue = 'payment_type', markers = True, ci=50)
res
```

```python
# The relplot shows the relationship between the purchases_partners and purchases and how they affect each other.

sns.relplot(x='purchases_partners', y='purchases', data=fintech, hue='housing')
```

```python
#The countplot shows the counts of observation from the different zodiac sign.

plt.figure(figsize=(15,5))
res = sns.countplot(x='zodiac_sign', data = fintech, palette = "Set1")
res
```

```python
plt.figure(figsize=(100,800))
sns.catplot(x='payment_type', y='reward_rate', data = fintech, hue = 'housing', jitter = False)
```
